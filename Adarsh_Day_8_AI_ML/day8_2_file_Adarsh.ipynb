```python
import matplotlib.image as mimg
import numpy as np
import matplotlib.pyplot as plt
path  = '/content/flower_image2.jpg'
im = mimg.imread(path)
print(im.shape)
plt.figure(1)
plt.imshow(im)
plt.axis('off')
```

```python
# rsize the image in 128x128
import cv2 # importing the computer vision library

newIm = cv2.resize(im,(128,128))
print(newIm.shape)
plt.figure(1)
plt.imshow(newIm)
plt.axis('off')
```

```python
from google.colab.patches import cv2_imshow
# seperate out all the color channel of the image
red = newIm[:,:,0] # first component
green = newIm[:,:,1] # second component
blue = newIm[:,:,2] # third component

cv2_imshow(red)
cv2_imshow(green)
cv2_imshow(blue)
```

```python
# create three new images array (numpy array)
# of uint8 type

rIM = np.zeros((newIm.shape[0],newIm.shape[1],newIm.shape[2]),
               dtype='uint8')
gIM = np.zeros((newIm.shape[0],newIm.shape[1],newIm.shape[2]),
               dtype='uint8')
bIM = np.zeros((newIm.shape[0],newIm.shape[1],newIm.shape[2]),
               dtype='uint8')
print(rIM.shape)

# replace the first component of rIM with red component
rIM[:,:,0]=red

# replace the second component of gIM with green component
gIM[:,:,1]=green

# replace the 3rd component of bIM with blue component
bIM[:,:,2]=blue


cv2_imshow(rIM)
cv2_imshow(gIM)
cv2_imshow(bIM)
```

```python
print(rIM.max(),rIM.min())
print(gIM.max(),gIM.min())
print(bIM.max(),bIM.min())
```
