```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt

# access any single image
usr_name = 14
samp_no = 6

path = "/content/orl_face/u%d/%d.png"%(usr_name,samp_no)

# read the image
im = mimg.imread(path)

print(type(im))
print(im.shape)
```

```python
# convert the 2D-data of image into 1-D features
feat = im.reshape(1,-1)# shift all the values of the image
#into the columns in one single row
print(im.shape)
print(feat.shape)
print("range:",im.min(),'-',im.max())
```

```python
#to access all the saples of the data
to_samp = 400
data = np.zeros((to_samp,im.shape[0]*im.shape[1]))
label = np.zeros((to_samp))
images = np.zeros((to_samp,im.shape[0],im.shape[1]))
indx = -1

for i in range(1,41,1): #traverse for user
   for j in range(1,11,1): #traverse for id
    indx = indx+1
    #access single image
    usr_name = i
    samp_no = j
    path = "/content/orl_face/u%d/%d.png"%(usr_name,samp_no)

    #read image
    im = mimg.imread(path)
    feat = im.reshape(1,-1)
    data[indx,:] = feat
    label[indx] = i
    images[indx,:,:] = im
    print("user num ",i,'samp no',j,'processed...')# display the image
```

```python
import numpy as np
from sklearn import svm
import pandas as pd
from sklearn import tree,svm,datasets,metrics,model_selection


X = data.copy()
y = label.copy()


# split the data into 70:30 ratio
Xtrain,Xtest,ytrain,ytest = model_selection.train_test_split(X,y,test_size=0.3,random_state=5)
print(Xtrain.shape,ytrain.shape)
print(Xtest.shape,ytest.shape)

ker = ['poly','linear','rbf']
c_value = [1,2,3]

# pre allocation of the result variable
result = np.zeros((len(ker),len(c_value)))
for i in range(len(ker)):
  for j in range(len(c_value)):
    # create the svm classifier
    orl_svm_model = svm.SVC(kernel=ker[i],gamma='scale',C=c_value[j])

    # train the model
    orl_svm_model = orl_svm_model.fit(Xtrain,ytrain)

    # predict the labels
    ypred = orl_svm_model.predict(Xtest)

    # accuracy
    acc = metrics.accuracy_score(ypred,ytest)
    #print("accuracy:", acc)
    result[i,j]=acc
print(result)

ResultDF = pd.DataFrame(result,index=ker,columns=["C=1","C=2","C=3"])
ResultDF
```

```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection

to_samp = 400
data = np.zeros((400,112,92))
label = np.zeros((to_samp))
indx = -1
for i in range(1,41,1): #traverse for user
   for j in range(1,11,1): #traverse for id
    indx = indx+1
    #access single image
    usr_name = i
    samp_no = j
    path = "/content/orl_face/u%d/%d.png"%(usr_name,samp_no)
    #read image
    im = mimg.imread(path)
    data[indx,:,:] = im
    label[indx] = i

(Xtrain,Xtest,ytrain,ytest) = model_selection.train_test_split(data,label,test_size=0.3,random_state=5)
print(Xtrain.shape,ytrain.shape)
print(Xtest.shape,ytest.shape)

print(len(np.unique(ytrain)))
```

```python
# create the CNN model
cnn_model = keras.models.Sequential() # empty framework

# Convolutinal layer 1
cnn_model.add(keras.layers.Conv2D(50,3,activation='relu',input_shape=(112,92,1)))

# maxpooling -1
cnn_model.add(keras.layers.MaxPool2D((2,2)))

# Convolutinal layer 2
cnn_model.add(keras.layers.Conv2D(50,3,activation='relu'))

# maxpooling -2
cnn_model.add(keras.layers.MaxPool2D((2,2)))

# feed forwards network
cnn_model.add(keras.layers.Flatten()) # input layer
cnn_model.add(keras.layers.Dense(200,activation='relu')) # HL1
cnn_model.add(keras.layers.Dense(200,activation='relu')) # HL2
cnn_model.add(keras.layers.Dense(200,activation='relu')) # HL3
cnn_model.add(keras.layers.Dense(400)) # Output layer

# optimizer
loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
cnn_model.compile(optimizer='sgd',loss = loss,metrics=['accuracy'])
cnn_model.summary()

# data scaling
Xtrain = Xtrain/Xtrain.max()
Xtest = Xtest/Xtest.max()

print(len(np.unique(ytrain)))
```

```python
# train the cnn along with the validation data
history = cnn_model.fit(Xtrain,ytrain,epochs=120,validation_data=(Xtest,ytest))
```
