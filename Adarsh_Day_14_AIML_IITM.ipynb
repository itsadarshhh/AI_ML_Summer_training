```python
import random
intents = {
            "greeting":["Hi! how are you?","Hello what's up"],
            "goodbye":["Great day ahead","Nice to see you"],
            "address":["Address of the college is Near Tihar Jail Janakpuri"],
            "courses":["IITM college offers 5 graduation courses and 2 postgraduation courses"],
            "Graduation":["graduation courses are : BCA, BBA , BCOM, Btech,BJMC"],
            "Postgraduation":["Post grad courses are: MBA, MCA"]
}
def detect_response(query):
  prediction = hb.Text.predict(
    text = query, # inferencing text
    labels=["greeting","goodbye",
            "address","courses",
            "Graduation","Postgraduation"] # suply potential labels
  )
  max_score=0
  for val in prediction:
    if(val['score']>max_score):
      max_score=val['score']
      max_val = val['score']
      max_val_text = val['className']
  print(max_val_text)
  return random.choice(intents[max_val_text])
```

```python
detect_response("Hi! how are you")
```

```python
import random
intents = {
            "greetings":["Hello! what would you like to have","Hi! how may i help you"],
            "pizza_options":["We have 4 pizza options: Margherita, Onion, Sweet Corn, Chicken"],
            "response":["Okay! anything else you wanna add on"],
            "add_ons":["We have additionals as well with pizza like: Softdrinks, Ice Cream, Fries, Double Cheese"],
            "offers":["We have combo pack","Buy 2 get 1 double cheese pan pizza option"],
            "costing":["This will cost you around XXXX"],
            "payment":["We offer 2 modes of transactetion: Cash, Online"],
            "waiting":["Sir/Ma'am your order number is XXX","Please wait your orderID is XXX","Your order will be ready in some time orderID:XXX"],
            "receiving":["Here's your order Thank You!!","Your order is ready THanks!"],
            "goodbye":["Great day ahead","Nice to see you"]
}
def detect_response(query):
  prediction = hb.Text.predict(
    text = query, # inferencing text
    labels=["greetings","pizza_options","response","add_ons","offers","costing","transaction","waiting","receiving","goodbye"] # suply potential labels
  )
  max_score=0
  for val in prediction:
    if(val['score']>max_score):
      max_score=val['score']
      max_val = val['score']
      max_val_text = val['className']
  print(max_val_text)
  return random.choice(intents[max_val_text])
```

```python
while(1):
  order = input()
  detect_response(order)
  if(order=="goodbye"):
    print(detect_response(order))
    break
  print(detect_response(order))
```
