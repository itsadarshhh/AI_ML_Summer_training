```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/ign.csv'
data = pd.read_csv(path)
print(data.shape)
data.head()
```

```python
ind = data['score_phrase']=='Disaster'
newdata = data[ind]
newdata[['title','platform']]
```

```python
# display the bar chart of score_phrase games
# in terms of  each score phrase occurance

plt.figure(1,figsize=(12,6))
ax = sns.countplot(data=data,x='score_phrase')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# display the bar chart of score_phrase games
# in terms of  each score phrase occurance with respect to editors_choice

plt.figure(1,figsize=(12,6))
ax = sns.countplot(data=data,x='score_phrase',hue = 'editors_choice')
for i in ax.containers:
  ax.bar_label(i)
```

```python
plt.figure(1,figsize=(12,6))
ax = sns.barplot(data=data,x='score_phrase',y='score',hue='editors_choice')
for i in ax.containers:
  ax.bar_label(i)
```

```python
path = '/content/Pokemon.csv'
data = pd.read_csv(path)
print(data.shape)
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
# remove the hash column
data.drop(['#'],axis=1,inplace=True) #it will remove the hash column and update the data variable itself
data.columns
```

```python
data.head()
```

```python
data.describe().astype(int)
```

```python
data['Attack'].describe().astype(int)
```

```python
# distribution plot of the attack

plt.figure(1,figsize=(12,6))
sns.displot(data=data,x='Attack',kind='hist')
```

```python
col = ['HP','Attack','Defense',	'Sp. Atk'	,'Sp. Def'	,'Speed'	,'Generation']
plt.figure(1,figsize=(15,10))
sns.boxplot(data=data[col])
```

```python
# distribution plot of attack , hp , defense , and speed in a single figure.
plt.subplot(2,2,1)
sns.histplot(data=data,x = 'Attack')
plt.subplot(2,2,2)
sns.histplot(data=data,x = 'HP')
plt.subplot(2,2,3)
sns.histplot(data=data,x = 'Speed')
plt.subplot(2,2,4)
sns.histplot(data=data,x = 'Defense')

```

```python
# diplay the count plot of type 1 pokemons

plt.figure(1,figsize=(12,6))
ax = sns.countplot(data=data,x='Type 1')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# display the count plot of type 2 pokemons.

plt.figure(1,figsize=(12,6))
ax = sns.countplot(data=data,x='Type 2')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# display the number of pokemons in each generation.
# show it in a visual form.

plt.figure(1,figsize=(12,6))
ax = sns.countplot(data=data,x='Generation')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# compare the type 1 pokemons with respect to attack capability.

plt.figure(1,figsize=(15,10))
ax = sns.barplot(data=data,x='Type 1',y = 'Attack')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# compare type 2 pokemons based upon their speed capabilities.

plt.figure(1,figsize=(15,10))
ax = sns.barplot(data=data,x='Type 2',y = 'Speed')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# rotate the labels at x axis.
plt.figure(1,figsize=(20,10))
sns.countplot(data=data,x='Generation')
plt.xticks(rotation=90)
```

```python
# compare type 2 pokemons generation wise based upon their speed capabilities.

plt.figure(1,figsize=(15,7))
ax = sns.barplot(data=data,x='Type 2',hue = 'Generation',y = 'Speed',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)
```

```python
# seperate out the legendary pokemons and non legendary pokemons
# then perform the analysis on legendary and non legendary pokemons separately

# compare the attack , defense and speed with respect to their types
# compare their total strength wrt to generation.

ind = data['Legendary']==True
newData = data[ind]
ind2 = data['Legendary']==False
newData2 = data[ind2]
print(newData.shape)
newData.describe()

```

```python
print(newData2.shape)
newData2.describe()
```

```python
# speed comparison in subfigures.
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,1)
ax = sns.barplot(data=newData,x='Type 1',y='Speed')
for i in ax.containers:
  ax.bar_label(i)
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,2)
ax = sns.barplot(data=newData2,x='Type 1',y='Speed')
for i in ax.containers:
  ax.bar_label(i)
```

```python
# Attack comparison in subfigures.
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,1)
ax = sns.barplot(data=newData,x='Type 1',y='Attack',errorbar= None)
for i in ax.containers:
  ax.bar_label(i)
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,2)
ax = sns.barplot(data=newData2,x='Type 1',y='Attack',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)
```

```python
# Defense comparison in subfigures.
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,1)
ax = sns.barplot(data=newData,x='Type 1',y='Defense',errorbar= None)
for i in ax.containers:
  ax.bar_label(i)
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,2)
ax = sns.barplot(data=newData2,x='Type 1',y='Defense',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)
```

```python
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,1)
sns.barplot(data=newData,x='Total',hue='Generation')
plt.figure(1,figsize=(15,7))
plt.subplot(1,2,2)
sns.barplot(data=newData2,x='Total',hue='Generation')
```

```python
# box representation of characteristics of pokemons

plt.figure(1,figsize=(15,7))
plt.subplot(1,2,1)
sns.boxplot(data=newData[col])
plt.title('Legendary Pokemons')
plt.subplot(1,2,2)
sns.boxplot(data=newData2[col])
plt.title('Non Legendary Pokemons')
```
